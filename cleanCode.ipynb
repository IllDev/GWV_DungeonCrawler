```python
from urllib.request import urlopen
import networkx as nx
import matplotlib.pyplot as plt

url = "https://nats-www.informatik.uni-hamburg.de/pub/GWV1920/%dcbungsaufgaben/blatt3_environment.txt"
array = []
arrayy = []
matrix = []
data = urlopen(url)

for line in data: 
    s = line.strip()
    array.append(s)

#matrix aufbauen aus dem vorhandenen array    
for elem in array:
    for char in elem:
        i = chr(char)
        arrayy.append(i) 
    matrix.append(arrayy)
    arrayy = []
```

```python
def neighbours(position):
    a, b = position
    speicher = []
    if (a+1, b) in validpos:
        i = a+1, b
        speicher.append(i)
    if (a-1, b) in validpos:
        i = a-1, b
        speicher.append(i)
    if (a, b+1) in validpos:
        i = a, b+1
        speicher.append(i)
    if (a, b-1) in validpos:
        i = a, b-1
        speicher.append(i)
    return speicher
```

```python
a = 0
b = 0 
dictionary = {}
validpos = []
dictionary2 = {}
for i in matrix:
    for j in i:
        if j == ' ':
            dictionary[a,b] = []
            i = a,b
            validpos.append(i)
        if j == 's':
            dictionary[a,b] = []
            i = a,b
            validpos.append(i)
        if j == 'g':
            dictionary[a,b] = []
            i = a,b
            validpos.append(i)
        a = a + 1
    a = 0
    b = b + 1

for key in dictionary.keys():
    a, b = key
    if (a, b) in validpos:
        dictionary2[key] = [neighbours(key)]
        #print(neighbours(key))
```

```python
Graph = nx.Graph()
for i in validpos:
    Graph.add_node(i)

for key in dictionary2:
    for var in dictionary2[key]:
        #print(var)
        for i in var:
            #print(i)
            Graph.add_edge(key, i)
```

```python
options = {
    'node_color': 'red',
    'node_size': 50,
    'line_color': 'grey',
    'linewidths': 0,
    'width': 0.3,
}
nx.draw(Graph, **options)
#plt.show()
```
